<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# X =
# y = 
X = df[['GRADES_8_G', 'EDUCATION SPENDING EFFECTIVENESS FACTOR', 'AVG_MATH_4_SCORE', 'AVG_READING_8_SCORE', 'YEAR']].dropna()


Y = df.loc[X.index]['AVG_MATH_8_SCORE']
Y.fillna(Y.median(), inplace=True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here

from sklearn.linear_model import LinearRegression
```

```python
# create your model here
# model = 

model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
model.score(X_test, Y_test)
```

```python
np.mean(model.predict(X_test) - Y_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_MATH_4_SCORE'

f = mplot.figure(figsize=(12,6))
mplot.scatter(X_train[col_name], Y_train, color = "red")
mplot.scatter(X_train[col_name], model.predict(X_train), color = "green")

mplot.legend(['True Training','Predicted Training'])
mplot.xlabel(col_name)
mplot.ylabel('AVG_READING_8_SCORE')
mplot.title("Model Behavior On Training Set")
```

```python
col_name = 'AVG_MATH_4_SCORE'

f = mplot.figure(figsize=(12,6))
mplot.scatter(X_test[col_name], Y_test, color = "blue")
mplot.scatter(X_test[col_name], model.predict(X_test), color = "black")

mplot.legend(['True testing','Predicted testing'])
mplot.xlabel(col_name)
mplot.ylabel('AVG_READING_8_SCORE')
mplot.title("Model Behavior on Testing Set")
```
